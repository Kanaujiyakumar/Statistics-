```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
#pip install matplotlib seaborn numpy pandas
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

size = 100
X = np.random.rand(size, 3) 
X[:, 1] = X[:, 0] * 0.5 + np.random.normal(0, 0.1, size)
X[:, 2] = X[:, 0] * 0.3 + np.random.normal(0, 0.2, size) 

df = pd.DataFrame(X, columns=['Feature1', 'Feature2', 'Feature3'])
corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, center=0, linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
```

```python
np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', linewidths=0.5)
plt.title('Variation in Sales Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
# 5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(25)

data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))

plt.scatter(df['GDP'], df['Population'], 
            s=df['Population'], 
            alpha=0.5,  
            c='blue', 
            edgecolors='w', 
            linewidth=2) 

for i, country in enumerate(df['Country']):
    plt.text(df['GDP'].iloc[i], df['Population'].iloc[i], country, 
             fontsize=9, ha='right', va='bottom')

plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs. GDP')
plt.grid(True)
plt.show()
```

```python
# 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

x = np.linspace(0, 10, 500)
y = np.sin(x)

output_notebook()

p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.line(x, y, legend_label="Sine Wave", line_width=2, color='blue')

show(p)
```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np
import pandas as pd

np.random.seed(42)
n = 100 
x = np.random.uniform(0, 10, n)
y = np.random.uniform(0, 10, n)
sizes = np.random.uniform(10, 100, n) 
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], n)

df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

output_notebook()

p = figure(title="Scatter Plot with Variable Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.scatter(x='x', y='y', source=df, size='sizes', color='colors', legend_field='colors', fill_alpha=0.6)

show(p)
```

```python
# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```

```python
from bokeh.plotting import figure, show, output_notebook
import pandas as pd

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

df = pd.DataFrame({'Fruit': fruits, 'Count': counts})

output_notebook()

p = figure(x_range=df['Fruit'], title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', width=800, height=400)

p.vbar(x='Fruit', top='Count', source=df, width=0.4, color='skyblue', legend_field='Fruit')
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.xaxis.major_label_orientation = "vertical"

show(p)
```

```python
# 4. Create a Bokeh histogram to visualize the distribution of the given data.
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

np.random.seed(42)
data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

output_notebook()

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='skyblue', line_color='black')

p.grid.grid_line_color = 'gray'
p.xaxis.axis_label_standoff = 12
p.yaxis.axis_label_standoff = 12

show(p)
```

```python
# 5. Create a Bokeh heatmap using the provided dataset.
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import numpy as np
import pandas as pd

# Generate random heatmap data
np.random.seed(42)
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

df = pd.DataFrame(data_heatmap, index=y, columns=x).reset_index().melt(id_vars='index')
df.columns = ['y', 'x', 'value']

output_notebook()

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=600, height=600,
           x_range=(0, 1), y_range=(0, 1), toolbar_location=None)

mapper = linear_cmap(field_name='value', palette=Viridis256, low=df['value'].min(), high=df['value'].max())

p.rect(x='x', y='y', width=0.1, height=0.1, source=df, line_color=None, fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

p.grid.grid_line_color = None
p.xaxis.major_label_orientation = 'horizontal'
p.yaxis.major_label_orientation = 'horizontal'

show(p)
```
